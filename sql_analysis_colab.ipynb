```python
import pandas as pd
import sqlite3

# Load your dataset
df = pd.read_csv("/content/CustomerChurn.csv")

# Create in-memory SQLite DB
conn = sqlite3.connect(":memory:")

# Load DataFrame into SQL
df.to_sql('churn_data', conn, index=False, if_exists='replace')
```

```python
print(df.columns)
```

```python
pd.read_sql_query('''
SELECT Churn, COUNT(*) as count
FROM churn_data
GROUP BY Churn
''', conn)
```

```python
pd.read_sql_query('''
SELECT "Senior Citizen", AVG("Monthly Charges") AS avg_day_minutes
FROM churn_data
GROUP BY "Senior Citizen"
''', conn)
```

```python
pd.read_sql_query('''
SELECT "Internet Service", Churn, COUNT(*) as count
FROM churn_data
GROUP BY "Internet Service", Churn
''', conn)
```

```python
pd.read_sql_query('''
SELECT "Customer ID", "Total Charges"
FROM churn_data
ORDER BY "Total Charges" DESC
LIMIT 5
''', conn)
```

```python
query = '''
SELECT "Customer ID", Churn
FROM churn_data
'''
df_sql = pd.read_sql_query(query, conn)
```

```python
df_sql.to_csv("sql_output_for_ml.csv", index=False)
```
